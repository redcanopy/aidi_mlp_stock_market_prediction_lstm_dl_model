#Random Forest Regressor and Decision Tree Regressor comparison

#Apple

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Pre_Processed_AAPL.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Days'] = (data['Date'] - data['Date'].min()).dt.days  # Create a numerical date feature

#Preprocess the data
features = ['Open', 'High', 'Low', 'Volume', 'Days']
target = 'Close'  # Target variable

X = data[features]
y = data[target]
dates = data['Date']  # Keep the Date column for plotting

# Split data into train and test sets
X_train, X_test, y_train, y_test, dates_train, dates_test = train_test_split(
    X, y, dates, test_size=0.2, random_state=42
)

# Train the models
# Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42, max_depth=5)
dt_model.fit(X_train, y_train)

# Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42, n_estimators=100, max_depth=5)
rf_model.fit(X_train, y_train)

# Make predictions
dt_predictions = dt_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)

# Calculate evaluation metrics
dt_mae = mean_absolute_error(y_test, dt_predictions)
dt_mse = mean_squared_error(y_test, dt_predictions)
dt_rmse = np.sqrt(dt_mse)
dt_mape = np.mean(np.abs((y_test - dt_predictions) / y_test)) * 100
dt_r2 = r2_score(y_test, dt_predictions)

rf_mae = mean_absolute_error(y_test, rf_predictions)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_rmse = np.sqrt(rf_mse)
rf_mape = np.mean(np.abs((y_test - rf_predictions) / y_test)) * 100
rf_r2 = r2_score(y_test, rf_predictions)

# Print out the results
print("Decision Tree Regressor Performance:")
print(f"RMSE: {dt_rmse:.4f}")
print(f"MAPE: {dt_mape:.4f}%")
print(f"R²: {dt_r2:.4f}")

print("\nRandom Forest Regressor Performance:")
print(f"RMSE: {rf_rmse:.4f}")
print(f"MAPE: {rf_mape:.4f}%")
print(f"R²: {rf_r2:.4f}")

# Sort by Date
results = pd.DataFrame({
    'Date': dates_test,
    'Actual': y_test,
    'DecisionTree': dt_predictions,
    'RandomForest': rf_predictions
}).sort_values(by='Date')

sorted_dates = results['Date']
actual_prices = results['Actual']
dt_predictions_sorted = results['DecisionTree']
rf_predictions_sorted = results['RandomForest']

# Visualize Predictions
plt.figure(figsize=(14, 7))
plt.plot(sorted_dates, actual_prices, label="Actual Prices", color='blue', linewidth=2)
plt.plot(sorted_dates, dt_predictions_sorted, label="Decision Tree Predictions", color='green', alpha=0.7)
plt.plot(sorted_dates, rf_predictions_sorted, label="Random Forest Predictions", color='red', alpha=0.7)
plt.legend()
plt.title("Stock Price Prediction: Linear Regression vs Decision Tree")
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.show()
```
